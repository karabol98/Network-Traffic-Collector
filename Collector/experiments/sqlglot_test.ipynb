```python
from sqlglot import parse_one, exp

# Original query
query = "SELECT * FROM users WHERE age > 20"

# Parse query into AST (Abstract Syntax Tree)
ast = parse_one(query)

# Modify LIMIT and OFFSET
# Set new LIMIT
ast.set("limit", exp.Limit(
    expression=exp.Literal.number(20)  # LIMIT value
))
ast.set("offset", exp.Offset(
    expression=exp.Literal.number(10)
))  # OFFSET value

# Print the modified query
print(ast.sql())
```
